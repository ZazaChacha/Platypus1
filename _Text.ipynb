As questions around gender equality remain a burning topic among the body politic and the electors, the global climate crisis intensifies.
A significant body of research demonstrated the existence of a gender gap in social and environmental behaviour. Women were shown to express more acute environmental concern,(Dunlap, Stern, Strapko). These behaviours appeared, among others, in situations of women leadership resulting in more efficient practice shifts toward sustainable businesses(Salamon), or in situations of women empowerment, to better allocation of income toward well-being of future generation (Duflo).

--It has been proven that women experience stronger the impact of climate-related damages than men (quote).
Climate change has more negative impact on women. Could women have a positive impact on climate change through (e.g) politics and policies?---

We have strong evidence that women participation in politics have a negative effect on aggravating factors of climate change such as CO2 emissions in a context of good political institutions (Lu). Building upon the latter result, this short research aims at examining the relationship between female parliamentarian representation and sustainable stringency at national scale, across Europe. and to determine whether  women's action at parliamentarian level (i.e in the highest political instances)  follow a similar pattern despite the political trends mentionned above.
